#Tensor Math Operatons

*   Add, Subtract, Multiply, Divide, Remainders, Exponents
*   Shorthand, Longhand
*   Reassignment





```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
tensor_a + tensor_b
```

```python
torch.add(tensor_a,tensor_b)
```

```python
tensor_b - tensor_a
```

```python
torch.sub(tensor_b, tensor_a)
```

```python
tensor_a * tensor_b
```

```python
torch.mul(tensor_a, tensor_b)
```

```python
tensor_b / tensor_a
```

```python
torch.div(tensor_b, tensor_a)
```

```python
tensor_b % tensor_a
```

```python
torch.remainder(tensor_b, tensor_a)
```

```python
tensor_a ** tensor_b
```

```python
torch.pow(tensor_a, tensor_b)
```

```python
# Another way
tensor_a.add(tensor_b)
```

```python
tensor_a
```

```python
# Using _ for reassignment (works only in this model)
tensor_a.add_(tensor_b)
```
