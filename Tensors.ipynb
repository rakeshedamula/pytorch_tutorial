#TENSORS

*   A torch.tensor is a multi-diemsional matrix containing elements of a single data type
*   Similar to Numpy Arrays, but works better on GPUs (vs regular CPUs)

*   Default data type is float32
*   More suitable for deep learning





```python
import torch
import numpy as np
```

#Lists

```python
my_list = [[1,2,3,4], [5,6,7,8]]
my_list
```

#Numpy Arrays

```python
np1 = np.random.rand(3,4)
np1
```

```python
np1.dtype
```

#Tensors

```python
tensor_2d = torch.randn(3,4)
tensor_2d
```

```python
tensor_2d.dtype
```

```python
tensor_3d = torch.zeros(2,3,4)
tensor_3d
```

```python
# Create tensor out of numpy array
tensor_np = torch.tensor(np1)
tensor_np
```
