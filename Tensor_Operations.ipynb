```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and view
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
# Reshape if we dont know the number of items by -1
my_torch2 = torch.arange(10)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(-1,5)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.view(2,5)
my_torch4
```

#Difference between Reshape and View

*   torch.reshape may return a copy or view of the original tensor. You cannot count on that to return a view or a copy.
*   If you need a copy use clone() if you need to use the same storage use view(). The semantics of reshape are that it may or may not share the storage and you dont know beforehand.

*   reshape() can operate on both contigious and non-contigious tensor while view() can operate only on contigious tensor
